1\. Write a SELECT statement that returns the same result set as this SELECT statement, but don’t use a join. Instead, use a subquery in a WHERE clause that uses the IN keyword.

SELECT DISTINCT vendor\_name

FROM vendors JOIN invoices

ON vendors.vendor\_id = invoices.vendor\_id ORDER BY vendor\_name

```sql
use AP;
```

```sql
select distinct VendorName
from Vendors
where VendorID IN
(select distinct VendorID
from Invoices)
order by VendorName;
```

Write a SELECT statement that answers this question: Which invoices have a payment total that’s greater than the average payment total for all invoices with a payment total greater than 0? Return the invoice_number and invoice_total columns for each invoice. This should return 20 rows. Sort the results by the invoice_total column in descending order.</span>

```sql
select InvoiceNumber, InvoiceTotal
from Invoices
where PaymentTotal>(select Avg(PaymentTotal) as average from Invoices where PaymentTotal<>0)
order by InvoiceTotal desc
```

Return one row for each account number that has never been assigned to any line item in the Invoice_Line_Items table. To do that, use a subquery intro-duced with the NOT EXISTS operator. This should return 54 rows.</span>

```sql
select distinct AccountNo, AccountDescription
from GLAccounts
where Not EXISTS (select * from InvoiceLineItems where InvoiceLineItems.AccountNo = GLAccounts.AccountNo);
```

5. <span style="color: rgb(0, 0, 0); font-family: arial, helvetica, sans-serif; font-size: 13.3333px; background-color: rgb(244, 244, 244);">What is average Sales quantity per order (use Sales order detail) which products have max quantity per order below overall average quantity. List product names for these products. Use CTE to make your query more readable</span>

```sql
with belowproduct AS
(
select ProductID, OrderQty 
from Sales.SalesOrderDetail as sod
where OrderQty<= (select avg(OrderQty) from Sales.SalesOrderDetail)
),
productnames AS
(
    select distinct Name, ProductID 
    from Production.Product as pp
)
select distinct Name, productnames.ProductID
from productnames join belowproduct 
on productnames.ProductID = belowproduct.ProductID;
```

```sql
--Question: to find average of listprice of each product (grouping by ProductID), which is greater than overall average of listprice
select ProductID, avg(ListPrice)
from Production.Product
Group by ProductID
HAVING avg(ListPrice) > (select avg(ListPrice) from Production.Product)
```
