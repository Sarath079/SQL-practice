1\. Write a SELECT statement that returns four columns based on the InvoiceTotal column of the Invoices table:  

- Use the CAST function to return the first column as data type decimal with 2 digits to the right of the decimal point.
-  Use CAST to return the second column as a varchar.
- Use the CONVERT function to return the third column as the same data type as the first column.
- Use CONVERT to return the fourth column as a varchar, using style 1.

```sql
use ap;
```

```sql
select cast(InvoiceTotal as decimal(7,2)) as deci, cast(InvoiceTotal as varchar) as '2nd column', 
 try_convert(decimal(7,2), InvoiceTotal) as convertdeci, convert(varchar, InvoiceTotal, 1) as '4th column'

from Invoices
```

2\. Write a SELECT statement that returns three columns based on the Invoice Date column of the Invoices table:  

- Use the CAST function to return the first column as data type varchar.  
- Use the CONVERT function to return the second and third columns as a varchar, using style 1 and style 10, respectively

```sql
select cast(InvoiceDate as varchar) as 'varchar', CONVERT(varchar, InvoiceDate, 1) as 'style1 varchar', CONVERT(varchar, InvoiceDate, 10) as 'style10 varchar'
from Invoices
```

3. <span style="color: rgb(0, 0, 0); font-family: arial, helvetica, sans-serif; font-size: 13.3333px; background-color: rgb(244, 244, 244);">Using tables Sales Order Detail and Product find out the top three products of class ‘M’ with the highest Gross Sales (Product Table column Class)</span>

```sql
use AdventureWorks2019;
```

```sql
select top 3 class, pro.ProductID, sum(OrderQty * UnitPrice) as gross
from Sales.SalesOrderDetail as sod join Production.Product as pro 
on sod.ProductID = pro.ProductID
where pro.class= 'M'
group by pro.ProductID, class
order by gross desc;
```

4. <span style="color: rgb(0, 0, 0); font-family: arial, helvetica, sans-serif; font-size: 13.3333px; background-color: rgb(244, 244, 244);">Using the results of query one, find out the top 3 customers for these products (you need to use a sub-query)&nbsp;</span>

```sql
SELECT top 3
    c.CustomerID, 
    p.Name, 
    SUM(sod.OrderQty * sod.UnitPrice) AS GrossSales
FROM 
    Sales.Customer AS c
    JOIN Sales.SalesOrderHeader AS soh ON c.CustomerID = soh.CustomerID
    JOIN Sales.SalesOrderDetail AS sod ON soh.SalesOrderID = sod.SalesOrderID
    JOIN (
        SELECT TOP 3 ProductID, Name, Class
        FROM Production.Product
        WHERE Class = 'M'
        ORDER BY (SELECT SUM(OrderQty * UnitPrice) FROM Sales.SalesOrderDetail WHERE Sales.SalesOrderDetail.ProductID = Production.Product.ProductID) DESC
    ) AS p ON sod.ProductID = p.ProductID
GROUP BY 
    c.CustomerID, 
    p.Name
ORDER BY 
    GrossSales DESC;
```
